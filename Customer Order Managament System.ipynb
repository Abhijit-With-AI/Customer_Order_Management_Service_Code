```python
import csv
import os
from datetime import datetime
import pandas as pd
```

```python
csv_file = 'order_details.csv'

if os.path.isfile(csv_file):
    pass
else:
    # Open a CSV file in write mode it create csv file
    with open('order_details.csv', 'w', newline='') as csvfile:
        # Create a CSV writer object
        csv_writer = csv.writer(csvfile)

        # Insert values into rows
        csv_writer.writerow(['Customer_ID', 'Name','Address','Order_details','Date&Time'])


while True:
    print("""
            1. Enter Customer Order
            2. See Customer Details(Using Customer_ID)
            3. Exit
    """)

    Choice=input("Enter which Operation You want to perform: ")

    if Choice=='1':
        while True:
            value=input("You want to enter new customer order details(y/n): ")
            if value=='y':
                # Get values to append
                Customer_ID = int(input("Enter Customer ID: "))
                data=pd.read_csv(csv_file)
                if Customer_ID in data['Customer_ID'].values:
                    print('Customer ID already exist.....')
                    break
                Name = input("Enter Customer Name: ")
                Address = input("Enter Customer Address: ")
                Order_details = input("Enter Customer Order: ")
                # Get the current date and time
                current_datetime = datetime.now()

                # Open the CSV file in append mode
                with open(csv_file, 'a', newline='') as csvfile:
                    # Create a CSV writer object
                    csv_writer = csv.writer(csvfile)

                    # Append values to a new row
                    new_values=[Customer_ID,Name,Address,Order_details,current_datetime]
                    csv_writer.writerow(new_values)
                    print("\nData Successfully store !!!!!!!!!!\n")
            else:
                break
    elif Choice=='2':
        data=pd.read_csv(csv_file)
        customer_id=int(input("Enter Your Customer ID Numder For Retrive The Information: "))
        if customer_id in data['Customer_ID'].values:
            target_row = data[data['Customer_ID'] == customer_id]
            print(target_row)
        else:
            print("There is no data available...")
    elif Choice=='3':
        break
    else:
        print("Enter The Right Choice!!!")
        
        
```
